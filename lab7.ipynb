```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from scipy import misc
import scipy


f=open('droptower_vdata.txt')
print f.readline()
print f.readline()
print f.readline()
print f.readline()
print f.readline()

#for line in f.readlines():
data=f.readlines()

for i in range(len(data)):
    data[i]=data[i][1:]
    data[i] = data[i].strip()
    

data[-1]=data[-1][1:]
for i in range(len(data)):
    data[i] = float(data[i])
print data

time=np.arange(0,11)
print time
f.close()
data_diff=np.diff(data)
time_diff = np.arange(1,11)
data_int= scipy.integrate.cumtrapz(data)
```

```python
plt.plot(time,data,marker='o',color='g')
plt.xlabel('time')
plt.ylabel('velocity')
plt.title('Velocity')
plt.savefig('part_1_velocity')
plt.show()


plt.plot(time_diff,data_diff,marker='o',color='b')
plt.xlabel('time')
plt.ylabel('acceleration')
plt.title('Acceleration')
plt.savefig('part_1_acceleration')
plt.show()


plt.plot(time_diff, data_int,marker='o',color='r')
plt.xlabel('time')
plt.ylabel('position')
plt.title('Position')
plt.savefig('part_1_position')
plt.show()
```

```python
plt.plot(time,data,marker='o',color='g', linestyle='dotted',label='Velocity')
plt.plot(time_diff,data_diff,marker='v',color='b', linestyle='dashed',label='Acceleration')
plt.plot(time_diff, data_int,marker='o',color='r',label='Position')
plt.title('A big ugly but presentable mess of values')
plt.xlabel('time')
plt.ylabel('all of them')
plt.xlim([0,14])
plt.ylim([-30,40])
plt.legend()
plt.savefig('part_2')
plt.show()
```

```python
f, (ax0,ax1,ax2) = plt.subplots(3, sharey=True)
ax0.plot(time,data,marker='o',color='g', linestyle='dotted',label='Velocity')
ax0.set_ylim([-30,20])
ax0.legend(loc='best')
ax1.plot(time_diff,data_diff,marker='v',color='b', linestyle='dashed',label='Acceleration')
ax1.set_ylim([-15,20])
ax1.legend(loc='best')
ax2.plot(time_diff, data_int,marker='o',color='r',label='Position')
ax2.set_ylim([-30,40])
ax2.legend(loc='best')
f.subplots_adjust(hspace=0.1)
ax0.set_title('all three plots in one image omg omg omg')
plt.savefig('part_3')
plt.show()
```
